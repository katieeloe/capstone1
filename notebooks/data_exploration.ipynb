```python
import numpy as np
import pandas as pd
pd.options.display.max_columns = 50
pd.options.display.max_colwidth = 80
from datetime import datetime
from datetime import timedelta
import scipy.stats as scs
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#read in text filed as pandas dataframes and remove spaces from column names
exam_df = pd.read_csv("data/ExamActivity_02062019_060002.txt", sep='|', parse_dates=True)
exam_df.columns = exam_df.columns.str.replace(' ', '_')
course_df = pd.read_csv("data/CourseActivity_02062019_060007.txt", sep='|', parse_dates=True)
course_df.columns = course_df.columns.str.replace(' ', '_')
course_df['Elapsed_Time'].replace(to_replace='0000:00:00.00', value='00:00:00', inplace=True)
```

```python
#convert date/time columns from strings to date/time objects
exam_df['Exam_Date'] =  pd.to_datetime(exam_df['Exam_Date'], format='%Y-%m-%d')
exam_df['Date_Recorded'] =  pd.to_datetime(exam_df['Date_Recorded'], format='%Y-%m-%d %H:%M:%S')
exam_df['Certificate_Expiration_Date'] =  pd.to_datetime(exam_df['Certificate_Expiration_Date'], format='%Y-%m-%d')
exam_df['Certifcate_Sent'] =  pd.to_datetime(exam_df['Certifcate_Sent'], format='%Y-%m-%d')
course_df['Date_Ordered'] =  pd.to_datetime(course_df['Date_Ordered'], format='%Y-%m-%d')
course_df['Date_Registered'] =  pd.to_datetime(course_df['Date_Registered'], format='%Y-%m-%d')
course_df['Expiration_Date'] =  pd.to_datetime(course_df['Expiration_Date'], format='%Y-%m-%d')
course_df['First_Login'] =  pd.to_datetime(course_df['First_Login'], format='%Y-%m-%d')
course_df['Last_Login'] =  pd.to_datetime(course_df['Last_Login'], format='%Y-%m-%d')
course_df['Date_Complete'] =  pd.to_datetime(course_df['Date_Complete'], format='%Y-%m-%d')
course_df['Elapsed_Time'] =  pd.to_timedelta(course_df['Elapsed_Time'])

#convert Pass/Fail column to 1 = pass & 0 = fail
exam_df['Exam_Status'] = exam_df['Exam_Status'].astype(str).str.strip()
exam_df['Exam_Status'].replace(to_replace=('Passed', 'Failed'), value=(1,0), inplace=True)
```

```python
#create dummy columns for course statuses 
course_status_dummies_df = pd.get_dummies(course_df['Course_Status'])
course_df = pd.concat([course_df, course_status_dummies_df], axis=1, ignore_index=False, copy=True)
course_df.columns = course_df.columns.str.replace(' ', '_')
```

```python
#filter out paper exam records
exam2 = exam_df[exam_df['Delivery_Type'] == 'Web']
```

```python
#filter out NULL EEID's
exam3 = exam2[pd.notna(exam2['EmployeeID'])]
#sort by Exam Date
exam4 = exam3.sort_values(by = 'Exam_Date')
```

```python
#group exams by EEID and select the earliest/first attempt
exam5 = exam4.loc[exam4.groupby('EmployeeID').Exam_Date.idxmin()]
```

```python
#filter out exams taken before 1/1/2017
exam_hypoth_df = exam5[exam5['Exam_Date'] >= '2017-01-01']
```

```python
#selecting courses with 10 or 11 sections complete as "course taken"
course_hypoth_df = course_df[(course_df['Current_Section'] > 9.0) & (course_df['Current_Section'] < 12.0)]
```

```python
#selecting columns from exam that I want to include in combined df
exam_col = ['NRA_Person_ID', 'Exam_Name', 'Exam_Form', 'Language', 'Exam_Date', 'Date_Recorded', 'Raw_Score', 
            'Pct_Score', 'Exam_Status','EmployeeID']
```

```python
#selecting columns from course that I want to include in combined df
course_col = ['NRA_Person_ID', 'Course_Name', 'Date_Registered', 'Expiration_Date', 'First_Login', 'Last_Login', 
              'Date_Complete', 'Current_Section', 'Course_Status', 'Elapsed_Time', 'Post_Test_Percent_Score', 
              'EmployeeID']
```

```python
#limiting df columnms on those selections
exam_hypoth_df = exam_hypoth_df[exam_col]
course_hypoth_df = course_hypoth_df[course_col]
```

```python
#merge exam and course dfs, inner join to allows each insatnce of course attributed to EEID to be lined up 
#with one instance (first attempt) for each EEID
exam_course_df = exam_hypoth_df.merge(course_hypoth_df, how = 'inner', on = 'EmployeeID')
exam_course_df.columns
```

```python
#create a time_delta column indicating the number of days bewteen course instance and exam instance
exam_course_df['time_delta'] = exam_course_df['Exam_Date'] - exam_course_df['Last_Login']
```

```python
#remove records with time_delta less than 0 (this indicates that the course was taken after the first exam attempt)
exam_course_df = exam_course_df[exam_course_df['time_delta'] >= pd.Timedelta(0, unit='d')]
```

```python
#sort values by time delta and reset index to allow us to select the course taken in closest proximity to the first exam
#attempt
exam_course_df = exam_course_df.sort_values('time_delta')
exam_course_df.reset_index(drop=True)
```

```python
#select the course taken in closest proximity to the first exam attempt
exam_course_df = exam_course_df.loc[exam_course_df.groupby('EmployeeID').time_delta.idxmin()]
```

```python
exam_course_df.describe()
```

```python
#select records only for employee who took the course less than week before the exam
exam_course_7d_df = exam_course_df[exam_course_df['time_delta'] <= pd.Timedelta(7, unit='d')]
exam_course_7d_df.shape
```

```python
#create list of EEIDs to remove from larger set of exam records
took_course_eeid = list(exam_course_7d_df['EmployeeID'])
```

```python
#create df of exam records where course was not taken, or taken more than 7 days prior to exam
exam_no_course_7d_df = exam_hypoth_df[~exam_hypoth_df.EmployeeID.isin(took_course_eeid)]
```

```python
exam_no_course_7d_df.shape
```

```python
#calculate total passes and attempts for each group to pass into a Beta distribution
took_course_passes = np.sum(exam_course_7d_df['Exam_Status'])
took_course_attempts = exam_course_7d_df.shape[0]
print(f"passed 1st attempts (course taken prior): {took_course_passes}")
print(f"total 1st attempts (course taken prior): {took_course_attempts}")
print(f"% passed 1st attempts (course taken prior): {took_course_passes/took_course_attempts}")
no_course_passes = np.sum(exam_no_course_7d_df['Exam_Status'])
no_course_attempts = exam_no_course_7d_df.shape[0]
print(f"passed 1st attempts (course NOT taken prior): {no_course_passes}")
print(f"total 1st attempts (course NOT taken prior): {no_course_attempts}")
print(f" % passed 1st attempts (course NOT taken prior): {no_course_passes/no_course_attempts}")
```

```python
exam_course_7d_dist = scs.beta(took_course_passes+1, took_course_attempts - took_course_passes +1)
exam_no_course_7d_dist = scs.beta(no_course_passes+1, no_course_attempts - no_course_passes +1)
```

```python
rv_samps = 10000
```

```python
took_course_samps = exam_course_7d_dist.rvs(rv_samps)
no_course_samps = exam_no_course_7d_dist.rvs(rv_samps)
prob_took_course_better = (took_course_samps > no_course_samps).mean()
print(prob_took_course_better)
print(took_course_samps)
print(no_course_samps)
```

```python
x1 = took_course_samps
x2 = no_course_samps
bins = np.linspace(.8, .925, 50)
plt.hist(x1, bins, alpha=0.5, label='took_course')
plt.hist(x2, bins, alpha=0.5, label='no_course')
plt.legend(loc='upper right')
plt.show()
```

```python
took_course_alpha = took_course_passes + 1
took_course_beta = took_course_attempts - took_course_passes + 1
no_course_alpha = no_course_passes + 1
no_course_beta = no_course_attempts - no_course_passes + 1
```

```python
took_course_7d_dist = scs.beta(took_course_alpha, took_course_beta)
no_course_7d_dist = scs.beta(no_course_alpha, no_course_beta)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
bins2 = np.linspace(0.8, 1.0, 201)

y_took_course = took_course_7d_dist.pdf(bins2)
y_no_course = no_course_7d_dist.pdf(bins2)

took_course_lines = ax.plot(bins2, y_took_course, label='Took Course within 7 Days of Exam', color = 'g')
ax.fill_between(bins2, y_took_course, alpha=0.2, color= 'g')

no_course_lines = ax.plot(bins2, y_no_course, label='Course Not Taken or Taken More than 7 Days Prior', color = 'r')
ax.fill_between(bins2, y_no_course, alpha=0.2, color= 'r')

plt.legend(loc='upper right')
plt.show()

```

```python
interval_size = 0.95
tail_area = (1 - interval_size) / 2

print("{0:0.1f}% credible interval for passing rate of employees who take the course within 7 days of taking the exam: {1:0.3f} (lower bound) - {2:0.3f} (upper bound)".format(
       interval_size *100,
       took_course_7d_dist.ppf(tail_area), 
       took_course_7d_dist.ppf(1 - tail_area)))
```

```python
#create Bayesian Test for course taken any number of days prior to exam
ever_took_course_eeid = list(exam_course_df['EmployeeID'])
no_course_ever_df = exam_hypoth_df[~exam_hypoth_df.EmployeeID.isin(ever_took_course_eeid)]
```

```python
#calculate total passes and attempts for each group to pass into a Beta distribution
ever_took_course_passes = np.sum(exam_course_df['Exam_Status'])
ever_took_course_attempts = exam_course_df.shape[0]
print(f"passed 1st attempts (course taken prior): {ever_took_course_passes}")
print(f"total 1st attempts (course taken prior): {ever_took_course_attempts}")
print(f"% passed 1st attempts (course taken prior): {ever_took_course_passes/ever_took_course_attempts}")
no_course_ever_passes = np.sum(no_course_ever_df['Exam_Status'])
no_course_ever_attempts = no_course_ever_df.shape[0]
print(f"passed 1st attempts (course NOT taken prior): {no_course_ever_passes}")
print(f"total 1st attempts (course NOT taken prior): {no_course_ever_attempts}")
print(f" % passed 1st attempts (course NOT taken prior): {no_course_ever_passes/no_course_ever_attempts}")
```

```python
ever_took_course_alpha = ever_took_course_passes + 1
ever_took_course_beta = ever_took_course_attempts - ever_took_course_passes + 1
no_course_ever_alpha = no_course_ever_passes + 1
no_course_ever_beta = no_course_ever_attempts - no_course_ever_passes + 1
```

```python
ever_took_course_dist = scs.beta(ever_took_course_alpha, ever_took_course_beta)
no_course_ever_dist = scs.beta(no_course_ever_alpha, no_course_ever_beta)
```

```python
fig, ax2 = plt.subplots(figsize=(10, 6))
bins3 = np.linspace(0.8, 1.0, 201)

y_took_course_ever = ever_took_course_dist.pdf(bins3)
y_no_course_ever = no_course_ever_dist.pdf(bins3)

ever_took_course_lines = ax2.plot(bins3, y_took_course_ever, label='Took Course at Some Point Prior to Exam', color = 'g')
ax2.fill_between(bins3, y_took_course_ever, alpha=0.2, color= 'g')

no_course_ever_lines = ax2.plot(bins3, y_no_course_ever, label='Course Not Taken Prior to Exam', color = 'r')
ax2.fill_between(bins3, y_no_course_ever, alpha=0.2, color= 'r')

plt.legend(loc='upper right')
plt.show()
```
